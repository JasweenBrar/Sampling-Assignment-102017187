### Import libraries

```python
import pandas as pd
import numpy as np
from collections import Counter
from imblearn.over_sampling import SMOTE

import math

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Load the dataset

```python
df = pd.read_csv("Creditcard_data.csv")
```

```python
df.head()
```

### Split the data into Input Features and Output Target

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
print("Original dataset shape is %s" % Counter(Y))
```

* We observe that dataset is highly imbalanced containing only 9 rows with Class:1 and 763 rows with Class:0

### So, let us perform SMOTE oversampling to balance the dataset

```python
sm = SMOTE(random_state=42)
X_new, Y_new = sm.fit_resample(X,Y)
```

```python
print("Resampled dataset shape is %s" % Counter(Y_new))
```

* Now the dataset is balanced

### SAMPLING TECHNIQUES

* Find out sample size and create sample

```python
data_balanced = pd.concat([X_new, Y_new], axis=1)
data_balanced.head()
```

#### 1) Simple Random Sampling

```python
# Set the confidence level
confidence_level = 0.95

# Set the margin of error
margin_of_error = 0.05

# Calculate the sample size
p = data_balanced['Class'].mean()  # proportion of the target class in the population
z = 1.96  # z-score for 95% confidence level
n = (z**2 * p * (1-p)) / (margin_of_error**2)
n = math.ceil(n)  # round up to the nearest integer
print("Sample size for random sampling:", n)

# Create the simple random sample
sample_simple_random = data_balanced.sample(n, random_state=42)
sample_simple_random.head()
```

#### 2) Stratified Sampling

```python
# Set the confidence level
confidence_level = 0.95

# Set the margin of error
margin_of_error = 0.05

# Calculate the sample size
p = data_balanced['Class'].mean()  # proportion of the target class in the population
z = 1.96  # z-score for 95% confidence level
n = (z**2 * p * (1-p)) / (margin_of_error**2)
n = math.ceil(n)  # round up to the nearest integer

strata = data_balanced.groupby('Class')

# sample 2 rows from each stratum
sample_stratified = strata.apply(lambda x: x.sample(n))

sample_stratified
N = len(sample_stratified)
print("Sample size for stratified sampling:", N)
sample_stratified.head()
```

#### 3) Cluster Sampling

```python
# Cluster Sampling
def get_clustered_Sample(df, n_per_cluster, num_select_clusters):
    N = len(df)
    K = int(N/n_per_cluster)
    data = None
    for k in range(K):
        sample_k = df.sample(n_per_cluster)
        sample_k["cluster"] = np.repeat(k,len(sample_k))
        df = df.drop(index = sample_k.index)
        data = pd.concat([data,sample_k],axis = 0)
    random_chosen_clusters = np.random.randint(0,K,size = num_select_clusters)
    samples = data[data.cluster.isin(random_chosen_clusters)]
    return(samples)
sample_cluster = get_clustered_Sample(df = data_balanced, n_per_cluster = 20, num_select_clusters = 28)

sample_cluster = sample_cluster.iloc[:,:-1]
N = len(sample_cluster)
print("Sample size for cluster sampling:", N)
sample_cluster.head()
```

#### 4) Systematic Sampling

```python
sample_systematic = data_balanced.iloc[::5]
## Systematically pick every 5th sample

N = len(sample_systematic)
print("Sample size for cluster sampling:", N)
sample_systematic.head()
```

#### 5) Multi-Stage Sampling

First perform cluster sampling and then randomly pick clusters

```python
# Cluster Sampling
def get_clustered_Sample(df, n_per_cluster, num_select_clusters):
    N = len(df)
    K = int(N/n_per_cluster)
    data = None
    for k in range(K):
        sample_k = df.sample(n_per_cluster)
        sample_k["cluster"] = np.repeat(k,len(sample_k))
        df = df.drop(index = sample_k.index)
        data = pd.concat([data,sample_k],axis = 0)
    random_chosen_clusters = np.random.randint(0,K,size = num_select_clusters)
    samples = data[data.cluster.isin(random_chosen_clusters)]
    return(samples)
sample_c = get_clustered_Sample(df = data_balanced, n_per_cluster = 20, num_select_clusters = 28)
sample_c = sample_c.iloc[:,:-1]

confidence_level = 0.95
margin_of_error = 0.05
# Calculate the random sample size
p = data_balanced['Class'].mean()  # proportion of the target class in the population
z = 1.96  # z-score for 95% confidence level
n = (z**2 * p * (1-p)) / (margin_of_error**2)
n = math.ceil(n)  # round up to the nearest integer

# Random sampling
sample_multistage = sample_c.sample(n , random_state=42)
N = len(sample_multistage)
print("Sample size for multi stage sampling:", N)
sample_multistage.head()
```

#### Apply 5 different models on all 5 samples
* DecisionTreeClassifier
* LogisticRegression
* GaussianNB
* GradientBoostingClassifier
* RandomForestClassifier

```python
# define the samples
samples = {"Sampling 1": sample_simple_random,
            "Sampling 2": sample_stratified,
            "Sampling 3": sample_cluster,
            "Sampling 4": sample_systematic,
            "Sampling 5": sample_multistage}


models = {'M1 = Decision Tree': DecisionTreeClassifier(random_state=0),
    'M2 = Logistic Regression': LogisticRegression(random_state=0),
    'M3 = Gaussian Naive Bayes': GaussianNB(),
    'M4 = Gradient Boosting Classifier': GradientBoostingClassifier(random_state=0),
    'M5 = Random Forest': RandomForestClassifier(random_state=0)}
```

```python
results=pd.DataFrame(columns= [sampling_name for sampling_name, _ in samples.items()], index= [model_name for model_name, _ in models.items()])

for sampling_name, sample in samples.items():
    for model_name, model in models.items():
        
        X_sample = sample.iloc[:,:-1]
        Y_sample = sample.iloc[:,-1]

        # Split the sample into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X_sample, Y_sample, test_size=0.2, random_state=42)

        # Train machine learning model on the training set
        model.fit(X_train, y_train)

        # Make predictions on the testing set
        y_pred = model.predict(X_test)

        # Calculate the accuracy score of the model
        accuracy = accuracy_score(y_test, y_pred)
        results.loc[model_name,sampling_name] = accuracy
```
